```python
import pandas as pd 
```

```python
ds =pd.read_csv('Restaurant_Reviews.tsv',delimiter='\t')
```

```python
ds
```

```python
Reviews = ds['Review']
```

```python
y=ds['Liked']
```

```python
import spacy
```

```python
english=spacy.load('en_core_web_sm')
```

```python
len(english.Defaults.stop_words)
```

```python
all_stopwords = english.Defaults.stop_words
```

```python
all_stopwords.remove('not')
```

```python

all_stopwords.remove('no')
```

```python
len(all_stopwords)
```

```python
import re
```

```python
corpus=[]
for line in Reviews :
    sentence = re.sub('[^a-zA-z]' , ' ' ,line)
    sentence = sentence.lower()
    tokens=english(sentence)
    stem_words =[]
    for word in tokens :
        if word.text not in all_stopwords :
            lem = word.lemma_
            stem_words.append(lem)
    stem_line = ' '.join(stem_words)
    corpus.append(stem_line)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bom = CountVectorizer(max_features=1000)
```

```python
X =bom.fit_transform(corpus)
```

```python
x=X.toarray()
```

```python
y=y.values
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
y_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train , y_train)
```

```python
y_predict =model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix ,accuracy_score
```

```python
confusion_matrix(y_test ,y_predict)
```

```python
accuracy_score(y_test , y_predict)
```
