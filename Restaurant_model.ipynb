```python
from nltk.corpus import stopwords
all_stopwords = stopwords.words('english')
```

```python
all_stopwords.remove("not")
all_stopwords.remove("no")
```

```python
len(all_stopwords)
```

```python
#corpus means colections of texts
corpus=[]
```

```python
for line in Review :
    stem_words=[]
    for word in line.split() :
        
        if word not in all_stopwords :
            lem_word = ss.stem(word)
            stem_words.append(lem_word)
    join_word = ' '.join(stem_words)
    corpus.append(join_word)
            
```
